# Probability
- Probability is the chance that an event will occur. Represented with numbers between 0 and 1.

- For example, in the case of flipping a coin (heads or tails).
- Each attempt to flip the coin is called a trial;
- Each result (heads or tails) is called an event;
- The sum of all possible events is called the sample space.

```python
import numpy as np
import pandas as pd
import scipy as stats
import seaborn as sns
```

## Permutation

- Is a counting technique used to determine how many ways there are to order the elements of a finite set.

```python
import math
```

```python
#In a simple permutation: 
#Getting the factorial to calculate the permutations between some variables, in this case 3 variables.

math.factorial(3)
```

- For example: if we have a total of 36 elements, a permutation of 5 elements of this set is (2,3,1,6,5). In this case, n = 36 and r = 5.
- Here is a way we can do that without repeting the numbers.

```python
math.factorial(36) / math.factorial(36-5)
```

```python
#When we can repeat the numbers:
math.pow(36, 5)
```

## Combination
- It is the possible number of arrangements in a beginning, without considering repetitions.

```python
math.factorial(6) / (math.factorial(2) * math.factorial(6-2))
```

- And if we consider repetitions:

```python
math.factorial(6+2-1)/(math.factorial(2)*math.factorial(6-1))
```

## Intersection, union and complement

```python
a = [0,1,2,3,4,5,6,7]
b= [0,2,4,6,8]
```

```python
#Intersection - Elements that are repeated are in both sets

set(a) & set(b)
```

Probability of intersection: Intersection / number of numbers that repeat in A and B.


```python
#Union - Is the union of the data of the two sets

set(a) | set(b)
```

Probability of joining the data of all sets = Prob(a) + Prob(b) - Intersection.


```python
#Difference - of a, only the elements who are in the A set

set(a).difference(set(b))
```

```python
#Complement - elements who are in one set but not in the other

set(a).symmetric_difference(set(b))
```

Complement of B with respect to A is = A-B


## Independent events
- The result of an event does not influence in the response of other event.
- For example, calculate the probability of getting two "tails" in two tries in a heads or tails.


## Dependent events
- the outcome of the first event influences the outcome of the second event. 
- For example: A deck has 52 cards and 13 of these are spades, what is the probability of drawing 2 spades?


## Conditional probability
- Calculate the probability of event A, given that event B has occurred.

# Probability and Binomial distribution

### Coins example

- Fliping the coin 10 times, wich is the probability of have 5 "heads"?


```python
n = 10 # amount of trials
x = 5  # number of successes
p = 0.5 # probability of sucess
```

```python
import math
(math.factorial(n)) / (math.factorial(x) * math.factorial(n - x)) * math.pow(p, x) * math.pow(1 - p, n - x)
```

```python
stats.binom.pmf(x, n, p)
```
