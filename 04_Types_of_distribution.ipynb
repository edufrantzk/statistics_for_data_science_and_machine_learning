## Bias (skew)
- When the distribution can be with the data concentrated to the left or to the right.


- If the data is biased, it may be that the collection was not done well. Only with specific groups as it does not follow a normal curve.

```python
from scipy.stats import skewnorm
```

```python
normal_d = skewnorm.rvs(a = 0, size = 1000)
```

```python
#Normal distribution

sns.distplot(normal_d);
```

```python
#Positive skew

normal_d_positive = skewnorm.rvs(a = 10, size = 1000)
sns.distplot(normal_d_positive);
```

```python
#Negative skew

normal_d_negative = skewnorm.rvs(-10, size = 1000)
sns.distplot(normal_d_negative);
```

```python
normal_d_negative.mean(), np.median(normal_d_negative), stats.mode(normal_d_negative)
```

## Standardized normal distribution
- With the standarization (z-score) the values are on the same scale, the mean is close to zero and the standard deviation is close to so. So it's better to compare different features.

```python
standard_normal_data = np.random.standard_normal(size=1000)
```

```python
min(standard_normal_data), max(standard_normal_data)
```

```python
sns.distplot(standard_normal_data);
```

```python
print(f'Mean:{standard_normal_data.mean()} Standard deviation:{np.std(standard_normal_data)}')
```

```python
data
```

```python
mean = data.mean()
std = np.std(data)
```

```python
print(mean)
print(std)
```

```python
#Z-Score
standardized = (data - mean) / std
standardized
```

```python
#The mean is zero and the Stdeviation is close to one

print(f'Mean: {standardized.mean()}, St Deviation: {np.std(standardized)}')
```

```python
sns.distplot(standardized);
```

### Central limit theory


- As the sample size increases, the sample distribution of its mean gets closer to a normal distribution.

```python
heights = np.random.randint(126, 192, 500)
heights.mean()
```

```python
#With just 3 heights

sns.distplot(heights);
```

```python
means = [np.mean(np.random.randint(126, 192, 500)) for _ in range(1000)]
```

```python
type(means), len(means)
```

```python
print(means)
```

```python
#With 1000 height
sns.distplot(means);
```

### Discrete variables

#### Bernoulli distribution
- Specific type of the binomial distribution.
- It has only two answers: success or failure.

```python
from scipy.stats import bernoulli
```

```python
bernoulli_data = bernoulli.rvs(size = 1000, p = 0.3) # p == probability
```

```python
bernoulli_data
```

```python
np.unique(bernoulli_data, return_counts=True)
```

```python
sns.distplot(bernoulli_data, kde = False) #kde = False wont show the kernel density estimation line
```

#### Binomial distribution
- Only two possibilities (success or failure).
- Probability of success or failure.
- In the binomial we can have more than two classes like single, married and divorced.

```python
from scipy.stats import binom
```

```python
binom_data = binom.rvs(size = 1000, n = 3, p = 0.8)
```

```python
np.unique(binom_data, return_counts=True)
```

```python
print(binom_data)
```

```python
sns.distplot(binom_data, kde=False);
```

### Poisson distribution


- Number of times an event happened in a time interval.
- Example: number of users who visited a website in an interval.
- Number of times an event happened in that interval (number of calls in a call center).

```python
from scipy.stats import poisson
```

```python
poisson_data_1 = poisson.rvs(size = 1000, mu = 10)
```

```python
min(poisson_data_1), max(poisson_data_1)
```

```python
np.unique(poisson_data_1, return_counts=True)
```

```python
sns.distplot(poisson_data_1, kde=False);
```

```python
poisson_data_2 = poisson.rvs(size = 1000, mu = 1)
```

```python
sns.distplot(poisson_data_2, kde=False)
```
